```python
import requests

def test_api(url, params=None, headers=None):
    print("Testing:", url)
    
    r = requests.get(url, params=params, headers=headers, timeout=30)
    
    print("Status Code:", r.status_code)
    print("Content-Type:", r.headers.get("Content-Type"))
    print("URL Used:", r.url)
    
    if r.status_code != 200:
        print("Error:", r.text[:300])
        return None
    
    try:
        data = r.json()
        print("JSON Type:", type(data))
        
        if isinstance(data, list):
            print("Number of Records:", len(data))
            print("Preview:", data[:2])
        elif isinstance(data, dict):
            print("Keys:", list(data.keys())[:10])
        
        return data
    
    except Exception:
        print("Non-JSON response")
        print(r.text[:300])
        return r.text

kiel = test_api(
    "https://v6.db.transport.rest/locations",
    params={"query": "Kiel Hbf"}
)
```

```python
import time
import requests

def test_api(url, params=None, headers=None, retries=5, backoff=2):
    """
    retries: wie oft wiederholen bei 429/503/502/504
    backoff: Sekunden, die jedes Mal größer werden (2, 4, 8, ...)
    """
    for attempt in range(1, retries + 1):
        try:
            print(f"\nTesting: {url} (attempt {attempt}/{retries})")
            r = requests.get(url, params=params, headers=headers, timeout=30)

            print("Status Code:", r.status_code)
            print("URL Used:", r.url)

            # Bei typischen "temporär" Fehlern warten und nochmal versuchen
            if r.status_code in (429, 502, 503, 504):
                wait = backoff ** (attempt - 1)
                print(f"Temporary error ({r.status_code}). Waiting {wait}s and retrying...")
                time.sleep(wait)
                continue

            print("Content-Type:", r.headers.get("Content-Type"))

            if r.status_code != 200:
                print("Error Preview:", r.text[:300])
                return None

            # JSON versuchen
            try:
                data = r.json()
                print("JSON Type:", type(data))
                if isinstance(data, list):
                    print("Number of Records:", len(data))
                    print("Preview:", data[:2])
                elif isinstance(data, dict):
                    print("Keys:", list(data.keys())[:10])
                return data
            except Exception:
                print("Non-JSON response preview:", r.text[:300])
                return r.text

        except requests.RequestException as e:
            wait = backoff ** (attempt - 1)
            print(f"Request failed: {e}. Waiting {wait}s and retrying...")
            time.sleep(wait)

    print("Failed after retries.")
    return None

kiel = test_api("https://v6.db.transport.rest/locations", params={"query": "Kiel Hbf"})
hamburg = test_api("https://v6.db.transport.rest/locations", params={"query": "Hamburg Hbf"})
```
