```python
import requests
import pandas as pd

def test_api(url, params=None):
    r = requests.get(url, params=params, timeout=30)
    print("Status:", r.status_code)
    print("URL:", r.url)
    print("Content-Type:", r.headers.get("Content-Type"))
    r.raise_for_status()
    return r

# Beispiel: Berlin, tägliche Mitteltemperatur
params = {
    "latitude": 52.52,
    "longitude": 13.41,
    "start_date": "2024-01-01",
    "end_date": "2024-01-31",
    "daily": "temperature_2m_mean",
    "timezone": "Europe/Berlin"
}

r = test_api("https://archive-api.open-meteo.com/v1/archive", params=params)
weather = r.json()

# in DataFrame
df_temp = pd.DataFrame({
    "date": weather["daily"]["time"],
    "temp_mean": weather["daily"]["temperature_2m_mean"]
})
df_temp["date"] = pd.to_datetime(df_temp["date"])
df_temp.head()
```

```python
import xml.etree.ElementTree as ET

xml_text = r.text

root = ET.fromstring(xml_text)

# Namespace aus Root holen (ACK hat ein anderes Namespace als Publication)
ns = {"ns": root.tag.split("}")[0].strip("{")}

reasons = root.findall(".//ns:Reason", ns)
if not reasons:
    print("No <Reason> element found. Full preview:")
    print(xml_text[:1500])
else:
    for reason in reasons:
        code = reason.findtext("ns:code", default="", namespaces=ns)
        text = reason.findtext("ns:text", default="", namespaces=ns)
        print("Reason code:", code)
        print("Reason text:", text)
```
