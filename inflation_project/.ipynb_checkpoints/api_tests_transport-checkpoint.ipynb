```python
import requests
import pandas as pd
from datetime import datetime
import io, zipfile
```

```python
# DB REALTIME API (RE70)
r_kiel = test_api("https://v6.db.transport.rest/locations", params={"query": "Kiel Hbf"})
kiel = r_kiel.json()
kiel[:3]

r_hh = test_api("https://v6.db.transport.rest/stations/search/Hamburg%20Hbf")
hamburg = r_hh.json()
hh_id = hamburg[0]["id"]

kiel_id, hh_id

r_j = test_api("https://v6.db.transport.rest/journeys", params={"from": hh_id, "to": kiel_id, "results": 6})
journeys = r_j.json()

# Kurz-Preview: Liniennamen in den Journeys suchen
def extract_lines(j):
    lines = []
    for leg in j.get("legs", []):
        line = leg.get("line") or {}
        name = line.get("name") if isinstance(line, dict) else str(line)
        if name:
            lines.append(name)
    return lines

for i, j in enumerate(journeys.get("journeys", [])[:6]):
    print(i, extract_lines(j))

re70_journeys = []
for j in journeys.get("journeys", []):
    lines = extract_lines(j)
    if any("RE70" in (ln or "") for ln in lines):
        re70_journeys.append(j)

print("Journeys with RE70:", len(re70_journeys))
re70_journeys[0] if re70_journeys else "No RE70 found in returned journeys"

deps_hh = test_api(f"https://v6.db.transport.rest/stops/{hh_id}/departures", params={"duration": 120}).json()
deps_kiel = test_api(f"https://v6.db.transport.rest/stops/{kiel_id}/departures", params={"duration": 120}).json()

def filter_departures_by_line(deps, target="RE70"):
    out = []
    for d in deps:
        line = d.get("line") or {}
        name = line.get("name") if isinstance(line, dict) else str(line)
        if name and target in name:
            out.append(d)
    return out

re70_hh = filter_departures_by_line(deps_hh, "RE70")
re70_kiel = filter_departures_by_line(deps_kiel, "RE70")

print("RE70 departures from Hamburg:", len(re70_hh))
print("RE70 departures from Kiel:", len(re70_kiel))

def show_fields(d):
    line = d.get("line") or {}
    return {
        "line": line.get("name") if isinstance(line, dict) else str(line),
        "direction": d.get("direction"),
        "plannedWhen": d.get("plannedWhen"),
        "when": d.get("when"),
        "delay_min": d.get("delay"),
        "platform": d.get("platform"),
        "cancelled": d.get("cancelled")
    }

show_fields(re70_hh[0]) if re70_hh else "No RE70 in HH departures"
```

```python
# Feiertage API für Schleswig-Holstein (DE) + Datum-Merge
holidays = test_api("https://date.nager.at/api/v3/PublicHolidays/2024/DE").json()
print("Records:", len(holidays))
holidays[0]
sh_holidays = []
for h in holidays:
    # counties kann None sein; wenn vorhanden: enthält z.B. "DE-SH"
    counties = h.get("counties")
    if counties is None or "DE-SH" in counties:
        sh_holidays.append(h)
print("Holidays relevant for SH:", len(sh_holidays))
```

```python
# OpenLigaDB — HSV & Holstein Kiel Events (Datum für Merge)
leagues = test_api("https://www.openligadb.de/api/getavailableleagues").json()
print("Leagues:", len(leagues))
leagues[:3]
# Beispielwerte – falls bei euch andere Keys auftauchen, passen wir sie an:
league_short = "bl2"   # häufig 2. Bundesliga
season = "2024"        # Saisonjahr

matches = test_api(f"https://www.openligadb.de/api/getmatchdata/{league_short}/{season}").json()
print("Matches:", len(matches))

# Filter HSV / Holstein Kiel
def is_team(m, name):
    t1 = (m.get("Team1") or {}).get("TeamName","")
    t2 = (m.get("Team2") or {}).get("TeamName","")
    return name.lower() in t1.lower() or name.lower() in t2.lower()

hsv_matches = [m for m in matches if is_team(m, "Hamburger SV")]
kiel_matches = [m for m in matches if is_team(m, "Holstein Kiel")]

print("HSV matches:", len(hsv_matches))
print("Holstein Kiel matches:", len(kiel_matches))

# Preview Datum
(hsv_matches[0].get("MatchDateTimeUTC"), hsv_matches[0].get("MatchDateTime")) if hsv_matches else "No HSV matches found"
```

```python
# CKAN package_show: liefert Metadaten + Ressourcen-URLs
pkg = test_api("https://opendata.schleswig-holstein.de/api/3/action/package_show", params={"id": "puenktlichkeit"}).json()
pkg["success"], list(pkg["result"].keys())
resources = pkg["result"]["resources"]
print("Resources:", len(resources))
for r in resources[:5]:
    print(r.get("name"), r.get("format"), r.get("url"))

# suche eine CSV/JSON Ressource
csv_url = None
for r in resources:
    fmt = (r.get("format") or "").lower()
    if "csv" in fmt:
        csv_url = r.get("url")
        break

print("CSV URL:", csv_url)

nahsh_df = pd.read_csv(csv_url)
print("Shape:", nahsh_df.shape)
print("Columns:", list(nahsh_df.columns))
nahsh_df.head(3)
# finde Linien-Spalte automatisch
line_candidates = [c for c in nahsh_df.columns if any(k in c.lower() for k in ["linie", "line", "route", "zug"])]
print("Line candidates:", line_candidates)
line_col = line_candidates[0] if line_candidates else None

re70_df = nahsh_df[nahsh_df[line_col].astype(str).str.contains("RE70", na=False)] if line_col else pd.DataFrame()
print("RE70 rows:", re70_df.shape)
re70_df.head(5)
date_candidates = [c for c in nahsh_df.columns if any(k in c.lower() for k in ["datum", "date", "monat", "month", "zeit"])]
print("Date candidates:", date_candidates)
date_col = date_candidates[0] if date_candidates else None

if date_col and not re70_df.empty:
    tmp = re70_df.copy()
    tmp[date_col] = pd.to_datetime(tmp[date_col], errors="coerce")
    print("Min:", tmp[date_col].min(), "Max:", tmp[date_col].max())
```

```python
def load_dwd_zip(url):
    r = test_api(url)
    r.raise_for_status()
    z = zipfile.ZipFile(io.BytesIO(r.content))
    files = [n for n in z.namelist() if n.lower().endswith((".csv",".txt"))]
    print("ZIP files:", files[:10])
    data_file = next((n for n in files if "produkt" in n.lower()), files[0])
    print("Using:", data_file)
    with z.open(data_file) as f:
        df = pd.read_csv(f, sep=";", engine="python")
    return df

DWD_HH_URL = "PASTE_DWD_HAMBURG_ZIP_URL_HERE"
DWD_KI_URL = "PASTE_DWD_KIEL_ZIP_URL_HERE"

dwd_hh = load_dwd_zip(DWD_HH_URL)
dwd_ki = load_dwd_zip(DWD_KI_URL)

print("Hamburg:", dwd_hh.shape, "Kiel:", dwd_ki.shape)
print("Columns HH:", list(dwd_hh.columns))
dwd_hh.head(3)
def date_range(df):
    date_cols = [c for c in df.columns if "DATUM" in c.upper() or "DATE" in c.upper()]
    print("Date cols:", date_cols)
    if date_cols:
        c = date_cols[0]
        df[c] = pd.to_datetime(df[c], errors="coerce")
        print("Min:", df[c].min(), "Max:", df[c].max())

date_range(dwd_hh)
date_range(dwd_ki)
```

```python
def html_keyword_check(url, keywords=("Hamburg","Kiel","RE70")):
    r = test_api(url, show_text_on_error=False)
    text = r.text if hasattr(r, "text") else ""
    found = {k: (k.lower() in text.lower()) for k in keywords}
    print("Keyword hits:", found)
    return found

html_keyword_check("https://www.bahn.de/service/fahrplaene/bauarbeiten")
html_keyword_check("https://bauprojekte.deutschebahn.com/")

test_api("https://developers.deutschebahn.com/db-api-marketplace/apis/product/isr-data-service")
```
